```python
import pandas as pd
import numpy as np
import os
import subprocess

# Make the graphs a bit prettier, and bigger
pd.set_option('display.line_width', 5000) 
pd.set_option('display.max_columns', 60) 
```

```python
video_map = pd.read_csv('/home/jingxi/hatch-video-map.csv')
```

```python
# run on raw interview data to generate clean groundtruth file and folder of corresponding videos
# target folder: '/dvmm-filer2/datasets/PersonalityVideos/Videos'
results = []
# count the number of hatch and role directories for each student
student_num_hatch = np.zeros(10000, int)
student_num_role = np.zeros(10000, int)

datapath = '/dvmm-filer2/datasets/PersonalityVideos/Interview'
incpath = '/dvmm-filer2/datasets/PersonalityVideos/Incomplete'
directories = os.listdir(datapath)
# for each directory
for directory in directories:
    if get_info(directory, video_map) == None:
        pass
    else:
        student_id, video_class, scores = get_info(directory, video_map)
        # print(student_id, video_class, scores)
        if video_class == 'hatch':
            student_num_hatch[student_id]+=1
            order = student_num_hatch[student_id]
        else:
            student_num_role[student_id]+=1
            order = student_num_role[student_id]
        folderpath = os.path.join(datapath, directory, 'video')
        videos = os.listdir(folderpath)
        # for each video
        for (i, video) in enumerate(videos):
            videopath = os.path.join(folderpath, video)
            entry_name = 'student' + str(student_id) + '_' + video_class + str(order) + '_video' + str(i+1)
            result = np.zeros(6, object)
            result[0] = entry_name
            result[1:6] = scores
            results.append(result)
            command = 'mv ' + videopath + ' /dvmm-filer2/datasets/PersonalityVideos/Videos/' + entry_name + '.mp4';
            subprocess.call(command, shell = True)
            print(command)
        
```

```python
ground_truths = pd.DataFrame(results, columns=['video', 'o', 'c', 'e', 'a', 'n'])
ground_truths.to_csv("ground_truths.csv")
```

```python
def get_info(directory, video_map):
    if video_map.loc[(video_map['hatch_video']==directory)].shape[0] != 0:
        video_class = 'hatch'
    elif video_map.loc[(video_map['role_video']==directory)].shape[0] != 0:
        video_class = 'role'
    else:
        # move this directory to '/dvmm-filer2/datasets/PersonalityVideos/Incomplete' if it has no entries (ground truth)
        source_path = os.path.join(datapath, directory)
        command = 'mv ' + source_path + ' ' + incpath ;
        print(command)
        subprocess.call(command, shell = True)
        return None
    student_id = video_map.loc[(video_map['hatch_video']==directory) | (video_map['role_video']==directory)].iloc[0]['student_id']
    scores = video_map.loc[video_map['student_id']==student_id].iloc[0]['o':'n'].values.astype(float)
    if any(np.isnan(scores)):
        # move this directory to '/dvmm-filer2/datasets/PersonalityVideos/Incomplete' if its entry has nan numbers
        source_path = os.path.join(datapath, directory)
        command = 'mv ' + source_path + ' ' + incpath ;
        print(command)
        subprocess.call(command, shell = True)
        return None
    return student_id, video_class, scores
```

```python
def process_video(videopath):
    path, videoname = os.path.split(videopath)
    destpath = '/dvmm-filer2/datasets/PersonalityVideos/Features/' + videoname.replace(".mp4","")
    mkdir_p(destpath)
    command = 'openface -f ' + videopath + ' -rigid' + ' -simalign -out_dir ' + destpath + ' -no2Dfp -no3Dfp -noMparams -noPose -noAUs -noGaze' + ' -q';
    print(command)
    subprocess.call(command, shell = True)
```

```python
def mkdir_p(path):
    try: 
        os.makedirs(path)
    except OSError as exc:  # Python >2.5
        if os.path.isdir(path):
            pass
        else:
            raise
```
